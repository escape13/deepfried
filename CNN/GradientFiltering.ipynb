```
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import urllib
```

```
image = urllib.request.urlretrieve('https://upload.wikimedia.org/wikipedia/commons/3/37/Theodore_Kaczynski.jpg', 'ted.jpg')
image = mpimg.imread('ted.jpg')
plt.imshow(image)
```

```
gray = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```
sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])
```

```
filtered_img = cv.filter2D(gray, -1, sobel_y)
plt.imshow(filtered_img, cmap='gray')
```

```
filtered_img = cv.filter2D(gray, -1, sobel_x)
plt.imshow(filtered_img, cmap='gray')
```
